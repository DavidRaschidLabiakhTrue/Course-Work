David: Relationship between X and Y with a cubic function


Y = aX^3 + bX^2 + cX + d

```python
X = XnonLinear
Y = YnonLinear

```

```python


learningRate = 0.000001
epochs = 10000

# coefficents
defaultCoefficient = 0
a = defaultCoefficient #coefficient
b = defaultCoefficient #coefficient
c = defaultCoefficient #coefficient
d = defaultCoefficient #coefficient

nX = float(len(X)) # length of X
summationConst = -2/nX


for i in range(epochs):
  Y_Pred = (a * X**3) + (b * X**2) + (c * X) + d
  difference = Y - Y_Pred # precompte the difference
  D_a = summationConst * sum(X**3 * (difference)) # deriv a
  D_b = summationConst* sum(X**2 * (difference))# deriv b
  D_c = summationConst * sum(X * (difference)) # deriv c
  D_d = summationConst * sum((difference)) # deriv d
  
  a = a - learningRate * D_a
  b = b - learningRate * D_b
  c = c - learningRate * D_c
  d = d - learningRate * D_d

Y_Pred = (a * X**3) + (b * X**2) + (c * X) + d
print("Optimal Solution:", a, b, c, d)




```

**Hints:**

David: Establishing Linear Function to describe relationship

Y1 = m1 * X1 + m2 * X2 + m3

```python
# Coding here

m1 = 0
m2 = 0
m3 = 0

learningRate = 0.001
epochs = 10000

leng = len(X1)
summationConst = -2/leng;

for i in range(epochs):
  Y_pred = (m1*X1) + (m2*X2) + (m3)
  D_m1 = summationConst * sum(X1 * (Y1 - Y_pred))
  D_m2 = summationConst * sum(X2 * (Y1 - Y_pred))
  D_m3 = summationConst * sum(Y1 - Y_pred)

  m1 = m1 - learningRate * D_m1
  m2 = m2 - learningRate * D_m2
  m3 = m3 - learningRate * D_m3

print("Optimal Solution:", m1, m2, m3)
```
